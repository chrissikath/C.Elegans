```python
import pandas as pd
```

```python
data = pd.read_csv("exon_coverage_annotation_merge.csv", delimiter="\t")
```

```python
onlycounts = data.iloc[:,1:-5] #get only counts
samples_names = data.iloc[:,0]
annotations = data.iloc[:,-5:]
column_list = list(onlycounts)
onlycounts["sum"] = onlycounts[column_list].sum(axis=1) #write sum of exons innerhalb sample
onlycountTPM = onlycounts.loc[:,:].div(onlycounts["sum"], axis=0) #divide each value with sum 
onlycountTPM *=1000000
onlycountTPM.insert(loc=0, column="sample",value=samples_names)
onlycountTPM = onlycountTPM.drop(['sum'], axis=1)
result = onlycountTPM.join(annotations, how='outer')
result.to_csv("exon_coverage_TPM.csv",index=False, sep="\t")
result
```
